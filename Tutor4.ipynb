  Secara umum hanya mencatat apakah kata tersebut muncul atau tidak di dalam dokumen 

2. TF-IDF

  TF: Term Frequency

  Frekuensi kemunculan suatu kata pada suatu dokumen

  IDF: Inverse Document Frequency

  Menghitung seberapa besar informasi yang bisa didapatkan dari suatu kata   pada semua dokumen yang ada

```python
texts = [
    "interaksi orang di facebook terhadap situs berita yang secara rutin mempublikasikan misinformasi melonjak",
    "facebook masih berjibaku menangkal peredaran hoaks dan berita keliru di platform-nya",
    "roda ekonomi seolah lumpuh dalam sekejap karena berbagai aktivitas dibatasi dalam beberapa bulan",
    "penambahan kasus terkonfirmasi positif virus corona jenis baru masih terjadi di berbagai negara",
    "jumlah pengguna berbayar layanan video on-demand disney plus terus tumbuh"
]
```

```python
#unique word
kata = sorted(set(word for sentence in texts for word in sentence.split()))
print(len(kata), kata)
```

```python
#mengubah ke vector
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(binary=True, token_pattern=r"(?u)\b\w+\b")
vec.fit(texts)
print([w for w in sorted(vec.vocabulary_.keys())])
```

```python
#menampilkan seluruh vector 
import pandas as pd
pd.DataFrame(vec.transform(texts).toarray(), columns=sorted(vec.vocabulary_.keys()))
```

```python
#Menghitung TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(stop_words='english')
vec.fit(texts)

import pandas as pd
pd.DataFrame(vec.transform(texts).toarray(), columns=sorted(vec.vocabulary_.keys()))
```
